```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
sns.set()
df_used_cars = pd.read_csv("bar_chart_data.csv")
df_used_cars.head(100)
df_used_cars["Cars Listings"]
```

```python
mlist=["#5c4d57","#815566","#a85b6b","#cc6464","#e97253","#fb8939","#ffa600"]
plt.figure(figsize = (8, 2))
plt.barh(y = df_used_cars["Brand"], width = df_used_cars["Cars Listings"], color = mlist)
plt.xticks(rotation = 0,fontsize = 10)
plt.yticks(rotation = 0, fontsize = 10)
plt.title("Cars Listings by Vladimr", fontsize = 12, fontweight = "bold")
plt.ylabel("[Number of Listings]", fontsize = 12 )
plt.savefig("Used Cars Bar.png")
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.facecolor'] = 'white'
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from matplotlib.ticker import PercentFormatter
df_real_estate = pd.read_csv("histogram_survey_data.csv")
df_real_estate.head(20)
```

```python
mlist=["#5c1627","#7d252e","#9e3831","#bc4f30","#d7692a","#ee861e","#ffa600"]

fig, ax = plt.subplots(figsize=(6, 6))
tick_size = 10
num_bins = np.arange(18, max(df_real_estate["Age"]) + 5, 5)

plt.title("Distribution of Age parameter", fontsize = 12, weight = "bold")
plt.xlabel("Age",fontsize=10)
plt.ylabel("Number of Properties",fontsize=10)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.facecolor'] = 'white'
n, bins, patches = plt.hist(df_real_estate["Age"], bins=num_bins, facecolor='#EEEEEE', edgecolor='#EEEEEE',
                            linewidth=0.75, alpha=0.5)
n = n.astype('int')
for i in range(len(num_bins) - 1):
    if n[i] > 0:
        ax.text(bins[i] + 2.5, n[i] + 1, str(int(n[i])), fontsize=10, ha='center', color='#7d252e')
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from matplotlib.ticker import PercentFormatter
df_spx_ftse_00_10 = pd.read_csv("returns1020.csv")
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
df_spx_ftse_00_10["new_date"]
```

```python
plt.figure(figsize = (20, 8))
labels = ["S&P 500", "FTSE 100"]
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.facecolor'] = 'white'
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2014-12-31') &
                                      (df_spx_ftse_00_10.new_date <= '2015-6-30')]

plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["GSPCRet"], color = "crimson")
plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["FTSERet"], color = "blue")
plt.title("S&P vs FTSE Returns from 2014 to 2015", fontsize = 14, fontweight = "bold") #updated here
plt.ylabel("Returns",fontsize = 12)
plt.xlabel("Date",fontsize = 12)
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
sns.set()
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
df_ad_expenditure
```

```python
sns.set(rc = {'figure.figsize': (8,6)})
sns.set_style("whitegrid", {"axes.facecolor": "1"})
sns.regplot(x = "Budget",
            y = "Sales", 
            data = df_ad_expenditure,
            scatter_kws = {'color': 'red'},
            line_kws = {'color': 'blue'})
plt.xlabel("Budget")
plt.ylabel("Sales")
plt.title("Graph Buget | Sales", fontsize = 12, weight = "bold")
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.facecolor'] = 'white'
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_print_revenue = pd.read_csv("area_chart_homework_data.csv")
df_print_revenue = df_print_revenue.rename(columns={'Unnamed: 4': 'Summary'})
df_print_revenue.head()
```

```python
df_print_revenue["Date"] = pd.to_datetime(df_print_revenue["Date"])
df_print_revenue["Summary"] =  df_print_revenue["International"] + df_print_revenue["Domestic"] + df_print_revenue["Services"]
df_print_revenue.head()
```

```python
labels = ["International", "Domestic", "Services"]
fig = plt.figure(figsize = (8, 6))
colors=["#bc4f30","#d7692a","#ee861e","#ffa600"]

plt.stackplot(df_print_revenue["Date"], df_print_revenue["International"], df_print_revenue["Domestic"], df_print_revenue["Services"], colors=colors)
#ax1 = fig.add_subplot(111)

plt.ylim(2000)
plt.legend(labels = labels, loc = "upper right")
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.rcParams['axes.facecolor'] = 'white'
plt.show()
```
